```python
import random

class StopExecution(Exception):
    def _render_traceback_(self):
        pass

class Creature():

    def __init__(self, name, hp=10, max_hp=10):
        self.name = name
        self.max_hp = max_hp
        if hp > max_hp:
            self.hp = max_hp
        else:
            self.hp = hp
        self.abilities = {"Attack": 1, "Defense": 5, "Speed": 2}

    def __str__(self):
        return f"Name: {self.name}, HP: {self.hp}, Max HP: {self.max_hp}, Abilities: {self.abilities}"

    def check_life(self):
        if self.hp <= 0:
            print(f"{self.name} fainted...")
            self.hp = 0
        return self.hp

    def attack(self, target):
        roll = random.randint(1, 20)
        print(f"{self.name} rolled a {roll}...")
        target_defence = target.abilities["Defense"] + target.abilities["Speed"]

        if roll < target_defence:
            print(f"{self.name}'s attack roll failed. {target.name}'s defence was too strong...")
        else:
            damage_done = random.randint(1, 4) + self.abilities["Attack"]
            target.hp -= damage_done
            print(f"{self.name}'s attack roll succeeded. {target.name} was hit for {damage_done} points of damage...")

    def turn(self, round_num, target):
        self.round_num = round_num
        self.attack(target)
        target.check_life()
        if target.hp <= 0:
            return False  # if they have fainted
        else:
            return True
        #print(f"{self.name}'s hp is {self.hp}.")
```

```python

class Fighter(Creature): 

    def __init__(self, name, hp=60, max_hp=50):
        super().__init__(name, hp)
        self.max_hp = 50
        if hp > max_hp:
            self.hp = self.max_hp
        else:
            self.hp = hp
        self.abilities = {"Attack": 5, "Defense": 10, "Speed": 3}
        self.count = 0

    def shield_up(self):
        if self.count == 0:
            self.abilities["Attack"] -= 5
            self.abilities["Defense"] += 5
            self.count += 1
            print(f"{self.name} is taking a defensive stance.")

    def shield_down(self):
        if self.count == 1:
            print(f"{self.name} is taking an attacking stance.")
            self.abilities["Attack"] += 5
            self.abilities["Defense"] -= 5
            self.count -= 1
            

    def turn(self, round_num, target):
        self.round_num = round_num
        print(f"{self.name} attacks {target.name}.")

        if round_num % 4 == 1:
            self.attack(target)
            target.check_life()
            if target.hp <= 0:
                self.shield_up()
                return False  # if they have fainted
            else:
                self.shield_up()
                return True

        if round_num % 4 == 2 or round_num % 4 == 3:
            self.attack(target)
            target.check_life()
            if target.hp <= 0:
                return False  # if they have fainted
            else:
                return True

        if round_num % 4 == 0:
            self.shield_down()
            self.attack(target)
            target.check_life()
            if target.hp <= 0:
                return False  # if they have fainted
            else:
                return True
```

```python
class Archer(Creature):

    def __init__(self, name, hp=50, max_hp=30):
        super().__init__(name, hp)
        self.max_hp = 30
        if hp > max_hp:
            self.hp = self.max_hp
        else:
            self.hp = hp
        self.abilities = {"Attack": 7, "Defense": 8, "Speed": 8}
        self.count = 0

    def sneak_attack(self, target):
        x = random.randint(1, 20)
        y = random.randint(1, 20)
        roll = max(x, y)

        if self.abilities["Speed"] > target.abilities["Speed"]:
            roll += self.abilities["Speed"] - target.abilities["Speed"]
        print(f"{self.name} performs a sneak attack. {self.name} rolled a {roll}...")

        if self.count == 0:
            self.abilities["Attack"] += 3
            self.abilities["Defense"] -= 3
            self.count += 1

        target_defence = target.abilities["Defense"] + target.abilities["Speed"]

        if roll < target_defence:
            print(f"{self.name}'s attack roll failed. {target.name}'s defence was too strong...")
        else:
            damage_done = random.randint(1, 8) + self.abilities["Attack"]
            target.hp -= damage_done
            print(f"{self.name}'s sneak attack roll succeeded. {target.name} was hit for {damage_done} points...")

    def attack(self, target):
        if self.count == 1:
            self.abilities["Attack"] -= 3
            self.abilities["Defense"] += 3
            self.count = 0
        roll = random.randint(1, 20)
        print(f"{self.name} rolled a {roll}...")
        target_defence = target.abilities["Defense"] + target.abilities["Speed"]

        if roll < target_defence:
            print(f"{self.name}'s attack roll failed. {target.name}'s defence was too strong...")
        else:
            damage_done = random.randint(1, 4) + self.abilities["Attack"]
            target.hp -= damage_done
            print(f"{self.name}'s attack roll succeeded. {target.name} was hit for {damage_done} points...")

    def turn(self, round_num, target):
        self.round_num = round_num
        print(f"{self.name} attacks {target.name}.")

        if self.round_num == 1 or self.round_num % 4 == 1:
            self.attack(target)
        else:
            self.sneak_attack(target)
        target.check_life()
        if target.hp <= 0:
            return False  # if they have fainted
        else:
            return True
```

```python
# Exercise 2 Task 1
class Goblin(Creature):

    def __init__(self, name, hp=15, max_hp=15):
        super().__init__(name, hp)
        self.max_hp = 15
        if hp > max_hp:
            self.hp = self.max_hp
        else:
            self.hp = hp
        self.abilities = {"Attack": 4, "Defense": 6, "Speed": 6}


class Orc(Creature):

    def __init__(self, name, hp=50, max_hp=50):
        super().__init__(name, hp)
        self.max_hp = 50
        if hp > max_hp:
            self.hp = self.max_hp
        else:
            self.hp = hp
        self.abilities = {"Attack": 10, "Defense": 6, "Speed": 2}
        self.count = 0

    def heavy_attack(self, target):
        roll = random.randint(1, 20)
        print(f"{self.name} rolled a {roll}...")
        target_defence = target.abilities["Defense"] + target.abilities["Speed"]

        if self.count == 0:
            self.abilities["Attack"] += 5
            self.abilities["Defense"] -= 3
            self.count += 1
            print(f"{self.name}'s attack increased and defense reduced.")

        if roll < target_defence:
            print(f"{self.name}'s heavy attack roll failed. {target.name}'s defence was too strong...")
        else:
            damage_done = random.randint(1, 4) + self.abilities["Attack"]
            target.hp -= damage_done
            print(f"{self.name}'s heavy attack roll succeeded. {target.name} was hit for {damage_done} points...")

    def attack(self, target):
        if self.count == 1:
            self.abilities["Attack"] -= 5
            self.abilities["Defense"] += 3
            self.count = 0
            print(f"{self.name}'s attack reduced and defense increased.")
        roll = random.randint(1, 20)
        print(f"{self.name} rolled a {roll}...")
        target_defence = target.abilities["Defense"] + target.abilities["Speed"]

        if roll < target_defence:
            print(f"{self.name}'s attack roll failed. {target.name}'s defence was too strong...")
        else:
            damage_done = random.randint(1, 4) + self.abilities["Attack"]
            target.hp -= damage_done
            print(f"{self.name}'s attack roll succeeded. {target.name} was hit for {damage_done} points...")

    def turn(self, round_num, target):
        self.round_num = round_num
        print(f"{self.name} attacks {target.name}.")

        if self.round_num % 4 == 0:
            self.heavy_attack(target)
        else:
            self.attack(target)
        print(f"{self.name}'s hp is {self.hp}.")
        target.check_life()
        if target.hp <= 0:
            return False  # if they have fainted
        else:
            return True
```

```python

class Wizard(Creature):
    def __init__(self, name, hp=20, max_hp=20, mana=100):
        self.name = name
        self.hp = max_hp
        self.max_hp = max_hp
        if hp > max_hp:
            self.hp = max_hp
        else:
            self.hp = max_hp
        self.abilities = {"Attack": 3, "Defense": 5, "Speed": 5, "Arcana": 10}
        self.mana = mana

    def audit_mana_level(self):
        if self.mana > 100:
            self.mana = 100
        if self.mana < 0:
            self.mana = 0
        print(f"mana count is {self.mana}")

    def audit_healing_level(self, target):
        if target.hp > target.max_hp:
            target.hp = target.max_hp

    def inward_audit_healing_level(self):
        if self.hp > self.max_hp:
            self.hp = self.max_hp

    def attack(self, target):
        self.mana += 20
        self.audit_mana_level()
        super().attack(target)

    def recharge(self):
        self.mana += 30
        self.audit_mana_level()

    def fire_bolt(self, target):
        arcana_addition = self.abilities["Arcana"] // 2
        roll = random.randint(1, 20) + arcana_addition
        print(f"{self.name} rolled a {roll}...")
        target_defence = target.abilities["Defense"] + target.abilities["Speed"]

        if roll < target_defence:
            print(f"{self.name}'s attack roll failed. {target.name}'s defence was too strong...")
        else:
            damage_done = random.randint(1, self.abilities["Arcana"])
            target.hp -= damage_done
            print(f"{self.name}'s attack roll succeeded. {target.name} was hit for {damage_done} points...")
            if self.mana < 90:
                self.mana += 10
                print(f"{self.name} gained 10 mana points as part of that attack...")
            else:
                print(f"{self.name} gained {100 - self.mana} mana points as part of that attack...")
                self.mana += 100 - self.mana

    def heal(self, target):
        if target == self.name and self.mana >= 20:
            self.mana -= 20
            self.hp += random.randint(0, 8) + self.abilities["Arcana"] // 2
            self.inward_audit_healing_level()
        elif self.mana >= 20:
            self.mana -= 20
            target.hp += random.randint(0, 8) + self.abilities["Arcana"] // 2
            self.audit_healing_level(target)
        else:
            print(f"{self.name} did not have enough mana to cast Heal on {target}")

    def mass_heal(self, allies):
        allies_list = []
        for i in allies:
            allies_list.append(i)

        if self.mana >= 30:
            self.mana -= 30
            mass_heal_value = random.randint(0, 10) + self.abilities["Arcana"]
            self.hp += mass_heal_value
            self.inward_audit_healing_level()
            print(f"{self.name} healed themselves for {mass_heal_value} points of HP...")
            for ally in allies:
                if ally.name != self.name:
                    mass_heal_value = random.randint(0, 10) + self.abilities["Arcana"]
                    print(f"{self.name} healed {ally.name} for {mass_heal_value} points of HP...")
                    ally.hp += mass_heal_value
                    self.audit_healing_level(ally)
        else:
            print(f"{self.name} did not have enough mana to cast Mass Heal.")

    def fire_storm(self, enemies):
        enemies_list = []
        for i in enemies:
            enemies_list.append(i)
        if self.mana >= 50:
            self.mana -= 50
            print(f"{self.name} used Fire Storm...")
            damage = random.randint(1, 20) + self.abilities["Speed"]
            if damage >= self.abilities["Arcana"]:
                self.hp -= damage // 2
            else:
                self.hp -= damage
            print(f"{self.name} was damaged for {damage} points of damage...")
            for i in enemies_list:
                damage = random.randint(5, 20) + self.abilities["Arcana"]
                print(f"{i.name} took {damage} points of damage...")
        else:
            print(f"{self.name} did not have enough mana and wasted their move.")
```

```python
class Battle():
    
    def __init__(self):
        self.battle_enemies = [GoblinKing("The Goblin King"), OrcGeneral("The Ork General"), Goblin("Goblino"), Orc("Orcington")]
        self.good_guys = [Fighter("Isildur"), Archer("Galadriel")]
        self.player = Wizard("Player")
        self.battle_allies = self.good_guys + [self.player]

    def enemy_cleanse(self):
        #print("FYI - checking enemies if any dead")
        for enemy in self.battle_enemies:
            if enemy.hp <= 0:
                self.battle_enemies.remove(enemy)
                #print(f"FYI - {enemy} removed from list of enemies")
        # for i in self.battle_enemies:
        #     print(i.name)

    def allies_cleanse(self):
       # print("FYI - checking allies if any dead")
        for ally in self.battle_allies:
            if ally.hp <= 0 and ally.name == self.player.name:
                battle.win_conditions()
            elif ally.hp <= 0:
                self.battle_allies.remove(ally)
                self.good_guys.remove(ally)
        #         print(f"FYI - {ally} removed from list of allies")
        # for i in self.battle_allies:
        #     print(i.name)
        # code commented out used for proofing purposes only.

    def auto_select(self, target_list):
        if not target_list:
            return None
        while True:
            target = random.choice(target_list)
            return target

    def select_target(self, target_list):
        if target_list:
            print("List of possible targets to attack: ")
            k = 1
            for target in target_list:
                print(f"{k}: {target.name}, HP: {target.hp}/{target.max_hp}")
                k += 1

            while True:
                try:
                    user = int(input("Enter choice: "))
                except ValueError:
                    print("Invalid choice. Please choose again: ")
                    continue
                else:
                    if user <= 0 or user > len(target_list):
                        continue
                break

            user -= 1
            return target_list[user]

    def win_conditions(self):
        if not self.battle_enemies:
            print("The good guys win!")
            quit()
        elif not self.good_guys:
            print("The bad guys win!")
            quit()
        elif self.player.hp <= 0:
            print("Player died and the forces of evil have won.")
            quit()
        else:
            return True

    def start(self):
        lineup = self.battle_enemies + self.battle_allies
        lineup = sorted(lineup, key=lambda x: x.abilities['Speed'], reverse=True)
        print("THE BATTLE BEGINS")
        round_num = 1
        while battle.win_conditions():
            print("===================================================")
            print(f"Round {round_num}.")
            print("===================================================")

            for i in lineup:
                if i.name == self.player.name:
                    battle.player_turn()
                    self.enemy_cleanse()
                    self.allies_cleanse()  # included in case Player burns themselves using Fire Storm
                    battle.win_conditions()
                elif self.good_guys and i in self.good_guys:
                    print(f"{i} turn to attack.")
                    targeted_creature = self.auto_select(self.battle_enemies)
                    i.turn(round_num, targeted_creature)
                    self.enemy_cleanse()
                    battle.win_conditions()
                    self.allies_cleanse()  # included in case Player burns themselves using Fire Storm
                    battle.win_conditions()
                elif self.battle_enemies and i in self.battle_enemies:
                    print(f"{i} turn to attack.")
                    targeted_creature = self.auto_select(self.battle_allies)
                    i.turn(round_num, targeted_creature)
                    self.allies_cleanse()
                    battle.win_conditions()

            print("===================================================")
            print(f"End of Round: {round_num}")
            print("===================================================")
            round_num += 1

        print("\nRoll credits......")

    def player_turn(self):
        print("===================================================\nPlayer turn")
        print("===================================================")
        print(f"Player: HP:{self.player.hp}:{self.player.max_hp} Mana: {self.player.mana}/100 \nAllies:")
        for i in self.good_guys:
            print(f"{i.name} HP: {i.hp}/{i.max_hp}")
        print("===================================================\nCurrent enemies:")
        for i in self.battle_enemies:
            print(f"{i.name} HP: {i.hp}/{i.max_hp}")
        print("===================================================")
        print("@ Player - Please pick an action:")
        print("Actions: F. Attack target (+20 mana)    R. Recharge (+30 mana)")
        print("Spells: 1. Fire Bolt (+10 mana on successful hit)     2. Heal (-20 mana)     3. Mass Heal (-30 mana)    4. Fire Storm (-50 mana)")
        print("Q. Quit Game")

        while True:
            player_move = str(input("Enter choice: ")).upper()

            if player_move == "F":
                self.player.attack(self.select_target(self.battle_enemies))
                break
            elif player_move == "R":
                self.player.recharge()
                break
            elif player_move == "1":
                self.player.fire_bolt(self.select_target(self.battle_enemies))
                break
            elif player_move == "2":
                self.player.heal(self.select_target(self.battle_allies))
                break
            elif player_move == "3":
                self.player.mass_heal(self.good_guys)
                break
            elif player_move == "4":
                self.player.fire_storm(self.battle_enemies)
                break
            elif player_move == "Q":
                print("Quitting game")
                raise StopExecution
            else:
                print("Invalid choice. Please choose again: ")
                continue
```
